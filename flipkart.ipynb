## Handling null values

```python
df.isnull().sum()
```

```python
df['Stock'].fillna(df['Stock'].median())
```

```python
df['Brand'].fillna('Unknown')
df
```

```python
df['Brand'].fillna('Unknown', inplace=True)
print(df.Brand)
```

```python
df['Ratings'].fillna(df['Ratings'].mean(), inplace=True)
```

```python
print(df.Ratings)
```

```python
df['Stock'].fillna(df['Stock'].mean(), inplace=True)
print(df.Stock)
```
